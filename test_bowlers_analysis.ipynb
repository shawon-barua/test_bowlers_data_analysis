```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv("wickets.csv", encoding = "ISO-8859-1")

display(df.head(10))
```

<br> Column Details
<br> Column Player :  Player Name 
<br> Span : Playing time
<br> Mat: Total Match played
<br> Inns : Total Innings he batted , every test match has two innings but sometime do not need to bat
<br> Balls : Total ball faced
<br> Runs : Total runs scored
<br> Wkts : Wicket taken
<br> BBI : Best bowling figure in a innings 
<br> BBM : Best bowling figure in a match
<br> Ave : Number of run given for per wicket
<br> Econ : Number of run given for per over
<br> SR : total balls bowled / wickets taken by bowler
<br> 5 : 5 wicket taken total
<br> 10 : 10 wicket taken total

```python
df = df.rename(columns={'Mat':'Match',
                        'Span': 'Playing Time',
                        'Wkts' : 'Total wickets taken',
                        'BBI': 'Best Bowling In a Innings',
                        'BBM' : 'Best Bowling In a Match',
                        'Inns':'Innings',
                        'HS': 'Highest_score',
                        'Ave': 'Average',
                        '10': '10 wicket match',
                        '5': '5 wicket match',
                        'Econ': 'Economy Rate',
                        'SR' : 'Strike Rate',
                       })

display(df.head())
```

```python
df.drop('Strike Rate', axis=1, inplace=True)
```
